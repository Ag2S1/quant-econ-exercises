```python
# Exercise 1

def factorial(n):
    return 1 if n == 0 else n * factorial(n-1)

factorial(10)
```

```python
# Exercise 2

%matplotlib inline

from math import factorial
import matplotlib.pyplot as plt

def binomial_rv(n, p):
    """"""
    y_values = []
    for k in range(0, n+1):
        y_values.append(factorial(n)/(factorial(k)*factorial(n-k))*pow(p,k)*pow(1-p, n-k))
    plt.plot(y_values, 'b+')
    plt.show()
    
binomial_rv(100,0.2)
```

```python
# Exercise 3

%matplotlib inline

from random import uniform
from math import sqrt

from __future__ import division
import matplotlib.pyplot as plt

N = 1000
area_count = 0
for i in range(0, N):
    x = uniform(0, 1)
    y = uniform(0, 1)
    if sqrt(x**2 + y**2) < 1:
        area_count += 1
        plt.plot(x, y, 'r+')
    else:
        plt.plot(x, y, 'b+')
plt.show()
pi = area_count / N * 4

print pi
```

```python
# Exercise 4

from random import uniform

N = 10
head_count = 0
pay = False
for i in range(0, N):
    if uniform(0, 1) >= 0.5:
        head_count += 1
        if head_count == 3:
            pay = True
            break
    else:
        head_count = 0
if pay:
    print "pay one dollar"
else:
    print "pay nothing"
```

```python
# Exercise 6

from pylab import plot, show, legend
from random import normalvariate

T = 200
alpha_values = [0.0, 0.8, 0.98]

def time_series(t, alpha):
    x = [0]
    for t in range(1, T):
        x.append(alpha*x[-1]+normalvariate(0, 1))
    return x
    
for alpha in alpha_values:
    plot(time_series(T, alpha), label='alpha='+str(alpha))
legend()
show()
```
